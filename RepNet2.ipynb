```python
# =============================Function Definitions=============================

clarity = {'FL-VVS2':'FL-VVS2', 'VS1-VS2':'VS1-VS2' ,  'SI1-SI2':'SI1-SI2'}
sizes = {  '0.30-0.39':clarity,
           '0.40-0.49':clarity,
           '0.50-0.69':clarity,
           '0.70-0.99':clarity,
           '1.00-1.49':clarity,
           '1.50-2.99':clarity}

shapes = {'Oval':sizes,
          'Pear':sizes,
          'Marquise':sizes,
          'Cushion (All)':sizes,
          'Emerald':sizes,
          'Princess':sizes,
          'Radiant':sizes,
          'Heart':sizes}

# Get current time
def get_current_time():
    current_time = time.time()
    return datetime.fromtimestamp(current_time)

start_program_time = get_current_time()

# SHAPE
def insert_shape_fn(text):
    element_present = EC.presence_of_element_located((By.XPATH, '//*[@id="classicSearchForm"]/div[3]/div[1]/div[1]/div[2]/div[1]/div/div[2]'))
    WebDriverWait(driver, wait_time).until(element_present)

    select_shape = driver.find_element(By.XPATH, '//*[@id="classicSearchForm"]/div[3]/div[1]/div[1]/div[2]/div[1]/div/div[2]')

    all_Shapes_Ele  = select_shape.find_elements(By.TAG_NAME, 'label')

    for sh in all_Shapes_Ele:
        if sh.text == text:
            sh.click()
            break
#     print("SHAPE DONE", end=", ")

# SIZE
def insert_size_fn(min_size, max_size):
    
    size_box = driver.find_element(By.XPATH,'/html/body/div[3]/div/main/section[2]/div/section/div/div/form/div[3]/div[1]/div[2]/div[2]')
    
    min_size_from = driver.find_element(By.XPATH, '//*[@id="filter.size.sizeFrom"]')
    to_max_size = driver.find_element(By.XPATH, '//*[@id="filter.size.sizeTo"]')
    
    min_size_from.send_keys(Keys.TAB)
    min_size_from.clear()
    
    min_size_from.send_keys(min_size)
    min_size_from.send_keys(Keys.TAB)
    to_max_size.send_keys(max_size)   

#     print("SIZE DONE", end=", ")

#COLOR
def insert_color_fn(color_from,to_color):
    color_form_ele = driver.find_element(By.XPATH,'//*[@id="classicSearchForm"]/div[3]/div[1]/div[3]')

    color_inbox_1 = color_form_ele.find_elements(By.TAG_NAME, 'input')[1]
    color_inbox_1.send_keys(color_from)
    color_inbox_1.send_keys(Keys.TAB)


    color_inbox_2 = color_form_ele.find_elements(By.TAG_NAME, 'input')[3]
    color_inbox_2.send_keys(to_color)
    color_inbox_2.send_keys(Keys.TAB)

#     print("COLOR D to K", end=", ")
    
# CLARITY
def insert_clarity_fn(clarity_from,to_clarity):

    clarity_form_ele = driver.find_element(By.XPATH,'//*[@id="classicSearchForm"]/div[3]/div[1]/div[4]')

    clarity_inbox_1 = clarity_form_ele.find_elements(By.TAG_NAME, 'input')[1]
    clarity_inbox_1.send_keys(clarity_from)
    clarity_inbox_1.send_keys(Keys.TAB)

    clarity_inbox_2 = clarity_form_ele.find_elements(By.TAG_NAME, 'input')[3]
    clarity_inbox_2.send_keys(to_clarity)
    clarity_inbox_2.send_keys(Keys.TAB)
```

```python
# INSERT SIZE
def  insert_size_fn(min_size, max_size):
    size_container = driver.find_element(By.XPATH,'/html/body/div[3]/div/main/section[2]/div/section/div/div[1]/div[2]/div/div[3]/div/form/div[3]/div[2]/div')
    size_fields = size_container.find_elements(By.TAG_NAME,'input')
    size_fields[0].clear()
    size_fields[0].send_keys(min_size)

    size_container = driver.find_element(By.XPATH,'/html/body/div[3]/div/main/section[2]/div/section/div/div[1]/div[2]/div/div[3]/div/form/div[3]/div[2]/div')
    size_fields = size_container.find_elements(By.TAG_NAME,'input')
    size_fields[1].clear()
    size_fields[1].send_keys(max_size)

# CLARITY
def insert_clarity_fn(clarity_from,to_clarity):
    clarity_filter = driver.find_element(By.XPATH,'/html/body/div[3]/div/main/section[2]/div/section/div/div[1]/div[2]/div/div[3]/div/form/div[5]/div[2]/div')

    clarity_inbox_1 = clarity_filter.find_elements(By.TAG_NAME, 'input')[0]
    clarity_inbox_1.clear()
    clarity_inbox_1.send_keys(clarity_from)
    clarity_inbox_1.send_keys(Keys.TAB)

    clarity_inbox_2 = clarity_filter.find_elements(By.TAG_NAME, 'input')[2]
    clarity_inbox_2.send_keys(to_clarity)
    clarity_inbox_2.send_keys(Keys.RETURN)
    clarity_inbox_2.send_keys(Keys.RETURN)
```
